
Algerian Forest Fires Dataset

Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

    Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)

Weather data observations 2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42 3. RH : Relative Humidity in %: 21 to 90 4. Ws :Wind speed in km/h: 6 to 29 5. Rain: total day in mm: 0 to 16.8 FWI Components 6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5 7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9 8. Drought Code (DC) index from the FWI system: 7 to 220.4 9. Initial Spread Index (ISI) index from the FWI

system: 0 to 18.5 10. Buildup Index (BUI) index from the FWI system: 1.1 to 68 11. Fire Weather Index (FWI) Index: 0 to 31.1 12. Classes: two classes, namely Fire and not Fire


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv('./data/Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

### Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```



The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region


```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df[df.isna().any(axis=1)]
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

### Remove null values

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
# drop 122 row which is header of second dataset
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# remove spaces in column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

### Change column valuse to int or float


```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
object = [feature for feature in df.columns if df[feature].dtype == 'O']
```

```python
for i in object:
    if i != "Classes":
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
## Save the cleaned dataset

df.to_csv('./data/Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
